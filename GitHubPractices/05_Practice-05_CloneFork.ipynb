## Process

Here are my codes and description

1. **Fork Existing repository using UI**
    - Created fork _(see repository https://github.com/inbrnthedj/gkpbt-css-circle.git)_ and exported it in a codespace to be made avaialble in the later steps by running:
```bash
    export ORIGIN=https://github.com/inbrnthedj/gkpbt-css-circle.git
```

2. **Clone forked repository in the lab environment**
    - clones the repository in current directory (in codespace)
```bash
    git clone $ORIGIN
```

3. **Create a new branch**
    - created a branch (-b) and switched to the branch created
```bash
    git checkout -b feature-circle-500
```

4. **Make changes locally**
    - I changed the height of the dimensions in style.css to 500px and checked the status, finding out that the change is not yet staged but Git is aware of the change:

        ```bash
            git status
        ```
    - To check the changes, I used git diff and it showed exactly where the change was, indicating with color red what was deleted and with green what was added. It showed basically in red the old lines and in green the new lines:

        ```bash
            git diff ./style.css
        ```

5. **Add and commit to local branch**
    - added all untracked files in the current directory and subdirectories to the staging area.
        ```bash
        git add .
        ```
    - "-A" if we add all files from the top level git folder
  

6. **Push changes to forked repository**
    - commit all the files in the staging area to the repository with a message
        ```bash
            git commit -m "Changing the height and the width of the circle"
        ```
    - then go back to the main branch to merge
        ```bash
            git checkout main
            git merge feature-circle-500
        ```
    - confirm changes by using git log command (short version)
        ```bash
            git log --oneline
        ```
    - ensure that I am on the main branch and delete the feature-circle-500 branch
        ```bash
            git checkout main
            git branch -d feature-circle-500
        ```
    - confirm if branch was deleted:
        ```bash
            git branch
        ```
    - finally, push all update to GitHub:
        ```bash
            git push origin main
        ```

7. **Create pull request to the upstream repository**
    - navigate to the forked repository, ensure you are on the Code tab. Click contribute and then Open pull request
    - it took me to "Comparing Changes" panel and showed the differences between the main branch of the fork and of the original repository.
    - Click Create pull request button
    - Add title and create pull request 
    - It told me that I've completed the lab.
